```python
from dateutil.parser import parse
import pandas as pd

def try_parse_datetime(string):
    try:
        return parse(string, dayfirst='dayfirst' in string) 
    except ValueError:
        return pd.NaT 

df['comment_date'] = df['comment_date'].apply(try_parse_datetime)
```

```python
df['comment_date'] = df['comment_date'].apply(lambda x: x.replace(tzinfo=None) if x else pd.NaT)
```

```python
earliest_datetime = df['comment_date'].min()

latest_datetime = df['comment_date'].max()

print(f"The dataset covers the period from {earliest_datetime} to {latest_datetime}.")
```
